## Data Preparation 

```python
user_identifier= df['user']
```

```python
df.drop(columns=['user'], inplace= True)
```

```python
df.columns.shape
```

```python
df= pd.get_dummies(df)
```

```python
df.columns
```

```python
df.drop(columns=['zodiac_sign_na','housing_na','payment_type_na'], inplace=True)
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split 
import random
```

```python
X_train,X_test,y_train,y_test= train_test_split(df.drop(columns=['churn']), df['churn'], train_size=0.7, random_state=0)
```

```python
X_train.shape
```

```python
y_train.value_counts()
```

```python
pos_index= y_train[y_train==1].index
neg_index= y_train[y_train==0].index
```

```python
if len(pos_index) > len(neg_index):
    higher= pos_index
    lower= neg_index
else:
    lower= pos_index
    higher= neg_index
```

```python
random.seed(0)
higher= np.random.choice(higher,size=len(lower))
lower= np.asarray(lower)
```

```python
new_indexes= np.concatenate((lower,higher))
```

```python
X_train= X_train.loc[new_indexes,]
y_train= y_train[new_indexes]
```

## feature scalling

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python

X_train2= pd.DataFrame(scaler.fit_transform(X_train))
X_test2= pd.DataFrame(scaler.transform(X_test))
X_train2.columns= X_train.columns.values
X_test2.columns= X_test.columns.values
X_train2.index= X_train.index.values
X_test2.index= X_test.index.values
X_train= X_train2
X_test= X_test2
```

```python
X_test
```

## Model 

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier= LogisticRegression(random_state=0)
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred= classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, precision_score, recall_score
```

```python
cm=confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

###  confusion matrix

```python
plt.figure(figsize= (15,10))
sns.heatmap(cm, annot= True)
```

###  crossvalidation

```python
from sklearn.model_selection import cross_val_score
accuracies= cross_val_score(estimator= classifier,
                           X=X_train,
                           y=y_train,
                           cv=10)
```

```python
accuracies
```

```python
accuracies.mean()
```
